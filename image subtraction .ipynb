```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
# Load the provided images
image_path_1 = 'resized_img/4_8_4_18_plane.png'
image_path_2 = 'resized_img/8_16_8_18_plane.png'
image_path_3 = 'resized_img/baseline_plane.png'

image1 = cv2.imread(image_path_1)
image2 = cv2.imread(image_path_2)
image3 = cv2.imread(image_path_3)

# Convert images to grayscale for subtraction
gray1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)
gray3 = cv2.cvtColor(image3, cv2.COLOR_BGR2GRAY)

# Perform image subtraction
diff_1_2 = cv2.absdiff(gray1, gray2)
diff_1_3 = cv2.absdiff(gray1, gray3)
diff_2_3 = cv2.absdiff(gray2, gray3)

# Display the results of the subtractions
plt.figure(figsize=(20, 10))

plt.subplot(1, 3, 1)
plt.title("Difference: 4_8_4_18 vs 8_16_8_18")
plt.imshow(diff_1_2, cmap='gray')
plt.axis("off")

plt.subplot(1, 3, 2)
plt.title("Difference: 4_8_4_18 vs Baseline")
plt.imshow(diff_1_3, cmap='gray')
plt.axis("off")

plt.subplot(1, 3, 3)
plt.title("Difference: 8_16_8_18 vs Baseline")
plt.imshow(diff_2_3, cmap='gray')
plt.axis("off")

plt.show()
```
